```python
import os

train_file = "UNSW_NB15_training-set.csv"
test_file = "UNSW_NB15_testing-set.csv"

# Check if files exist
if not os.path.exists(train_file):
    print(f"‚ùå Train file not found: {train_file}")
else:
    print(f"‚úÖ Train file found: {train_file}")

if not os.path.exists(test_file):
    print(f"‚ùå Test file not found: {test_file}")
else:
    print(f"‚úÖ Test file found: {test_file}")
```

```python
import pandas as pd

# Load train and test data
try:
    train_data = pd.read_csv(train_file, on_bad_lines="skip", low_memory=False)
    print("‚úÖ Train data loaded successfully!")
except Exception as e:
    print(f"‚ùå Error loading train data: {e}")

try:
    test_data = pd.read_csv(test_file, on_bad_lines="skip", low_memory=False)
    print("‚úÖ Test data loaded successfully!")
except Exception as e:
    print(f"‚ùå Error loading test data: {e}")
```

```python
print("\nüîç Missing Values in Training Data:")
print(train_data.isnull().sum())

print("\nüîç Missing Values in Test Data:")
print(test_data.isnull().sum())
```

```python
print("\nüìÑ Training Data Columns:")
print(train_data.columns)

print("\nüìÑ Test Data Columns:")
print(test_data.columns)

# Check if columns match
if set(train_data.columns) == set(test_data.columns):
    print("‚úÖ Column names match between train and test datasets!")
else:
    print("‚ùå Column names do not match between train and test datasets!")
```

```python
print("\nüìä Training Data Label Distribution:")
print(train_data["label"].value_counts())

print("\nüìä Test Data Label Distribution:")
print(test_data["label"].value_counts())

print("\nüìä Training Data attack_cat Distribution:")
print(train_data["attack_cat"].value_counts())

print("\nüìä Test Data attack_cat Distribution:")
print(test_data["attack_cat"].value_counts())
```

```python
print("\nüîç Duplicates in Training Data:", train_data.duplicated().sum())
print("üîç Duplicates in Test Data:", test_data.duplicated().sum())

# Remove duplicates (if necessary)
train_data = train_data.drop_duplicates()
test_data = test_data.drop_duplicates()
```

```python
print("\nüìä Training Data Describe:")
print(train_data.describe())

print("\nüìä Test Data Describe:")
print(test_data.describe())
```

```python
print("\nüîç Corrupted Rows in Training Data:")
print(train_data.isna().any(axis=1).sum())

print("\nüîç Corrupted Rows in Test Data:")
print(test_data.isna().any(axis=1).sum())
```

```python
# Check for overlapping rows (if unique identifiers exist)
if "id" in train_data.columns and "id" in test_data.columns:
    overlap = set(train_data["id"]).intersection(set(test_data["id"]))
    print("\nüîç Overlapping Rows Between Train and Test:", len(overlap))
else:
    print("\nüîç No unique identifier column found to check for overlapping rows.")
```

```python
print("\n‚úÖ Final Verification Summary:")
print(f"Training Data Shape: {train_data.shape}")
print(f"Test Data Shape: {test_data.shape}")
print(f"Training Data Columns: {train_data.columns.tolist()}")
print(f"Test Data Columns: {test_data.columns.tolist()}")
print(f"Training Data Label Distribution:\n{train_data['label'].value_counts()}")
print(f"Test Data Label Distribution:\n{test_data['label'].value_counts()}")
print(f"Training Data attacks types Distribution:\n{train_data['attack_cat'].value_counts()}")
print(f"Test Data attacks types Distribution:\n{test_data['attack_cat'].value_counts()}")
```

```python
# Remove overlapping rows from the test set based on the 'id' column
test_data = test_data[~test_data["id"].isin(train_data["id"])]

# Verify the shapes after removing overlaps
print(f"Training Data Shape: {train_data.shape}")
print(f"Test Data Shape: {test_data.shape}")

# Check if there are still overlapping rows
overlap = set(train_data["id"]).intersection(set(test_data["id"]))
print(f"\nüîç Overlapping Rows Between Train and Test: {len(overlap)}")
```

```python
# Define a mapping for attack categories to unique numbers
def map_attack_category(attack_cat):
    if attack_cat == 'Normal':
        return 'Normal'
    elif attack_cat in ['DoS', 'Exploits', 'Fuzzers', 'Generic']:
        return 'DoS'
    elif attack_cat == 'Reconnaissance':
        return 'Probe'
    elif attack_cat in ['Analysis', 'Backdoors']:
        return 'R2L'
    elif attack_cat in ['Shellcode', 'Worms']:
        return 'U2R'
    else:
        return 'Unknown'

train_data['attack_cat']=train_data['attack_cat'].apply(map_attack_category)
test_data['attack_cat']=test_data['attack_cat'].apply(map_attack_category)
# Verify the encoding

print("\nüìä Training Data Attack Categories Distribution:")
print(train_data['attack_cat'].value_counts())

print("\nüìä Test Data Attack Categories Distribution:")
print(test_data['attack_cat'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()
train_data['attack_cat']=label_encoder.fit_transform(train_data['attack_cat'])
test_data['attack_cat']=label_encoder.transform(test_data['attack_cat'])
```

```python
lb_proto=LabelEncoder()
train_data['proto']=lb_proto.fit_transform(train_data['proto'])
test_data['proto']=lb_proto.transform(test_data['proto'])
```

```python
lb_service=LabelEncoder()
train_data['service']=lb_service.fit_transform(train_data['service'])
test_data['service']=lb_service.transform(test_data['service'])
```

```python
lb_state=LabelEncoder()
train_data['state']=lb_state.fit_transform(train_data['state'])
test_data['state']=lb_state.transform(test_data['state'])
```

```python


# Verify the new columns
print("\nüìÑ Training Data Columns After Encoding:")
print(train_data.columns)

print("\nüìÑ Test Data Columns After Encoding:")
print(test_data.columns)
```

```python
from sklearn.preprocessing import StandardScaler

# Define numerical columns (excluding 'id', 'label', and 'attack_cat')
numerical_cols = train_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols.remove('id')  # Remove 'id' column
numerical_cols.remove('label')  # Remove 'label' column
numerical_cols.remove('attack_cat')  # Remove 'attack_cat' column

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform on training data
train_data[numerical_cols] = scaler.fit_transform(train_data[numerical_cols])

# Transform test data using the scaler fitted on training data
test_data[numerical_cols] = scaler.transform(test_data[numerical_cols])

# Verify normalization
print("\nüìä Training Data Describe (After Normalization):")
print(train_data[numerical_cols].describe())

print("\nüìä Test Data Describe (After Normalization):")
print(test_data[numerical_cols].describe())
```

```python
# Features (X) - All columns except 'id', 'label', and 'attack_cat'
X_train = train_data.drop(columns=['id', 'label', 'attack_cat'])
X_test = test_data.drop(columns=['id', 'label', 'attack_cat'])

# Labels (y) - The 'attack_cat' column
y_train = train_data['attack_cat']
y_test = test_data['attack_cat']

# Verify shapes
print("\nüìä Training Data Shapes:")
print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")

print("\nüìä Test Data Shapes:")
print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import learning_curve
import numpy as np
import time
```

```python

# --- Visualization: Distribution of Target Variable ---
plt.figure(figsize=(10, 6))
sns.countplot(x=y_train, order=y_train.value_counts().index)
plt.title('Distribution of Attack Categories in Training Data')
plt.xlabel('Attack Category (Encoded)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=y_test, order=y_test.value_counts().index)
plt.title('Distribution of Attack Categories in Test Data')
plt.xlabel('Attack Category (Encoded)')
plt.ylabel('Count')
plt.show()
```

```python
print("\n--- Training KNN Model ---")
knn = KNeighborsClassifier(n_neighbors=5) # You can tune n_neighbors

start_time = time.time()
knn.fit(X_train, y_train)
end_time = time.time()
print(f"KNN Training Time: {end_time - start_time:.2f} seconds")
```

```python
# --- KNN Model ---

start_time = time.time()
y_pred_knn = knn.predict(X_test)
end_time = time.time()
print(f"KNN Prediction Time: {end_time - start_time:.2f} seconds")

print("\n--- KNN Results ---")
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
```

```python
# Visualization: KNN Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test, y_pred_knn), annot=True, fmt='d', cmap='Blues')
plt.title('KNN Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Visualization: KNN Learning Curve (Optional, can take time)
print("\n--- Generating KNN Learning Curve ---")
train_sizes, train_scores, test_scores = learning_curve(knn, X_train, y_train, cv=3, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 5))

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
plt.title("KNN Learning Curve")
plt.xlabel("Training Examples")
plt.ylabel("Score")
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python
X_train_rnn_lstm = X_train.values.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_rnn_lstm = X_test.values.reshape((X_test.shape[0], 1, X_test.shape[1]))
```

```python
# Convert labels to one-hot encoding for multi-class classification
from tensorflow.keras.utils import to_categorical
num_classes = len(np.unique(y_train))
y_train_one_hot = to_categorical(y_train, num_classes=num_classes)
y_test_one_hot = to_categorical(y_test, num_classes=num_classes)
```

```python

print(f"\nRNN/LSTM Input Shape (Train): {X_train_rnn_lstm.shape}")
print(f"RNN/LSTM Input Shape (Test): {X_test_rnn_lstm.shape}")
print(f"RNN/LSTM Label Shape (Train): {y_train_one_hot.shape}")
print(f"RNN/LSTM Label Shape (Test): {y_test_one_hot.shape}")
```

```python
# --- RNN Model ---
print("\n--- Training RNN Model ---")
rnn_model = Sequential()
rnn_model.add(SimpleRNN(64, return_sequences=False, input_shape=(X_train_rnn_lstm.shape[1], X_train_rnn_lstm.shape[2])))
rnn_model.add(Dropout(0.2))
rnn_model.add(Dense(num_classes, activation='softmax'))

rnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Use early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

start_time = time.time()
history_rnn = rnn_model.fit(X_train_rnn_lstm, y_train_one_hot, epochs=20, batch_size=32, validation_split=0.2, callbacks=[early_stopping], verbose=0)
end_time = time.time()
print(f"RNN Training Time: {end_time - start_time:.2f} seconds")
```

```python

print("\n--- RNN Results ---")
start_time = time.time()
loss_rnn, accuracy_rnn = rnn_model.evaluate(X_test_rnn_lstm, y_test_one_hot, verbose=0)
end_time = time.time()
print(f"RNN Evaluation Time: {end_time - start_time:.2f} seconds")

print(f"Test Accuracy: {accuracy_rnn:.4f}")

y_pred_rnn_prob = rnn_model.predict(X_test_rnn_lstm)
y_pred_rnn = np.argmax(y_pred_rnn_prob, axis=1)

print("Classification Report:\n", classification_report(y_test, y_pred_rnn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rnn))
```

```python
# Visualization: RNN Training History
plt.figure(figsize=(12, 6))
plt.plot(history_rnn.history['accuracy'], label='Train Accuracy')
plt.plot(history_rnn.history['val_accuracy'], label='Validation Accuracy')
plt.title('RNN Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(history_rnn.history['loss'], label='Train Loss')
plt.plot(history_rnn.history['val_loss'], label='Validation Loss')
plt.title('RNN Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```

```python


# Visualization: RNN Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test, y_pred_rnn), annot=True, fmt='d', cmap='Blues')
plt.title('RNN Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python

# --- LSTM Model ---
print("\n--- Training LSTM Model ---")
lstm_model = Sequential()
lstm_model.add(LSTM(64, return_sequences=False, input_shape=(X_train_rnn_lstm.shape[1], X_train_rnn_lstm.shape[2])))
lstm_model.add(Dropout(0.2))
lstm_model.add(Dense(num_classes, activation='softmax'))

lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Use early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

start_time = time.time()
history_lstm = lstm_model.fit(X_train_rnn_lstm, y_train_one_hot, epochs=20, batch_size=32, validation_split=0.2, callbacks=[early_stopping], verbose=0)
end_time = time.time()
print(f"LSTM Training Time: {end_time - start_time:.2f} seconds")
```

```python
print("\n--- LSTM Results ---")
start_time = time.time()
loss_lstm, accuracy_lstm = lstm_model.evaluate(X_test_rnn_lstm, y_test_one_hot, verbose=0)
end_time = time.time()
print(f"LSTM Evaluation Time: {end_time - start_time:.2f} seconds")
print(f"Test Accuracy: {accuracy_lstm:.4f}")

y_pred_lstm_prob = lstm_model.predict(X_test_rnn_lstm)
y_pred_lstm = np.argmax(y_pred_lstm_prob, axis=1)

print("Classification Report:\n", classification_report(y_test, y_pred_lstm))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lstm))
```

```python


# Visualization: LSTM Training History
plt.figure(figsize=(12, 6))
plt.plot(history_lstm.history['accuracy'], label='Train Accuracy')
plt.plot(history_lstm.history['val_accuracy'], label='Validation Accuracy')
plt.title('LSTM Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(history_lstm.history['loss'], label='Train Loss')
plt.plot(history_lstm.history['val_loss'], label='Validation Loss')
plt.title('LSTM Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()

# Visualization: LSTM Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test, y_pred_lstm), annot=True, fmt='d', cmap='Blues')
plt.title('LSTM Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
